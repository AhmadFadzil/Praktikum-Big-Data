```python
pipeline = [
    {
        '$sort': {
            'salary': -1  # Mengurutkan karyawan berdasarkan gaji tertinggi
        }
    },
    {
        '$group': {
            '_id': {
                'department': '$department',  # Mengelompokkan berdasarkan departemen
                'employee_id': '$_id'  # Memastikan ID karyawan tetap
            },
            'name': {'$first': '$name'},  # Mengambil nama karyawan
            'salary': {'$first': '$salary'}  # Mengambil gaji karyawan
        }
    },
    {
        '$sort': {
            '_id.department': 1,  # Mengurutkan berdasarkan nama departemen
            'salary': -1  # Mengurutkan berdasarkan gaji tertinggi dalam setiap departemen
        }
    },
    {
        '$group': {
            '_id': '$_id.department',  # Mengelompokkan kembali berdasarkan departemen
            'employees': {
                '$push': {
                    'employee_id': '$_id.employee_id',
                    'name': '$name',
                    'salary': '$salary'
                }
            }
        }
    },
    {
        '$project': {
            'employees': {
                '$slice': ['$employees', 5]  # Mengambil 5 karyawan teratas dari setiap departemen
            }
        }
    }
]

# Menjalankan agregasi dan menampilkan hasil
results = collection.aggregate(pipeline)

# Mencetak hasil dengan format yang lebih rapi
for result in results:
    print(f"Departemen: {result['_id']}")
    print("5 Karyawan dengan Gaji Tertinggi:")
    for emp in result['employees']:
        print(f"  - Nama: {emp['name']}, Gaji: {emp['salary']}")
    print()  # Menambahkan baris kosong untuk pemisah antar departemen
```

```python
from pymongo import MongoClient

# Inisialisasi client dan koneksi ke database
client = MongoClient('mongodb+srv://Fadzil:Fadzil32@cluster0.8dewj.mongodb.net/')
db = client['company_db']
collection = db['employees']

# Menghapus karyawan yang berusia di bawah 25 tahun
result = collection.delete_many({'age': {'$lt': 25}})

# Menampilkan jumlah karyawan yang dihapus
print(f'Jumlah karyawan yang dihapus: {result.deleted_count}')
```

```python
pipeline = [
    {
        '$group': {
            '_id': '$department',  # Mengelompokkan berdasarkan departemen
            'total_salary': {'$sum': '$salary'},  # Menghitung total gaji
            'average_age': {'$avg': '$age'}  # Menghitung rata-rata umur
        }
    },
    {
        '$project': {
            'department': '$_id',  # Menyimpan nama departemen
            'total_salary': 1,  # Menyimpan total gaji
            'average_age': 1,  # Menyimpan rata-rata umur
            '_id': 0  # Menghilangkan _id default dari output
        }
    },
    {
        '$sort': {
            'department': 1  # Mengurutkan hasil berdasarkan nama departemen
        }
    }
]

# Menjalankan agregasi dan menampilkan hasil
results = collection.aggregate(pipeline)

# Mencetak laporan
print("Laporan Gaji dan Umur Karyawan per Departemen:")
for result in results:
    average_age = result['average_age']  # Mengambil rata-rata umur
    average_age_display = average_age if average_age is not None else 0.00  # Mengatur nilai default jika None

    print(f"Departemen: {result['department']}, Total Gaji: {result['total_salary']}, Rata-rata Umur: {average_age_display:}")
```
