```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Membaca data dari file CSV
df = pd.read_csv('mcu_box_office.csv')


# Menghapus koma (,) dan mengonversi kolom production_budget menjadi tipe numerik
df['production_budget'] = df['production_budget'].str.replace(',', '').astype(float)

# Menampilkan dataset sebelum normalisasi
print("Dataset Sebelum Normalisasi:")
print(df)

# Normalisasi kolom numerik 'production_budget'
scaler = MinMaxScaler()

# Melakukan normalisasi pada kolom 'production_budget'
df[['production_budget']] = scaler.fit_transform(df[['production_budget']])

# Menampilkan dataset setelah normalisasi
print("\nDataset Setelah Normalisasi:")
print(df)

# Menyimpan hasil normalisasi ke dalam file CSV baru (opsional)
df.to_csv('file_data_normalized.csv', index=False)  # Menyimpan ke file baru
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca data dari file CSV
df = pd.read_csv('file_data_normalized.csv')

# Membuat plot menggunakan seaborn
plt.figure(figsize=(8, 10))

# Visualisasi scatter plot antara tomato_meter dan audience_score
sns.scatterplot(x='tomato_meter', y='audience_score', data=df, hue='movie_title', palette='Set1', s=100)

# Menambahkan judul dan label
plt.title('Relationship between Tomato Meter and Audience Score', fontsize=16)
plt.xlabel('Tomato Meter (%)', fontsize=12)
plt.ylabel('Audience Score (%)', fontsize=12)

# Menampilkan plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca data dari file CSV
df = pd.read_csv('file_data_normalized.csv')

# Membuat bar plot untuk tomato_meter
plt.figure(figsize=(8, 5))
sns.barplot(x='tomato_meter', y='movie_title', data=df, palette='Set2')

# Menambahkan judul dan label
plt.title('Tomato Meter by Movie', fontsize=16)
plt.xlabel('Tomato Meter (%)', fontsize=12)
plt.ylabel('Movie Title', fontsize=12)

# Menampilkan plot
plt.show()
```
